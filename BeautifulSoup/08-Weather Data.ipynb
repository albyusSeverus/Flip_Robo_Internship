### Write a python program to extract information about the local weather from the National Weather Servicewebsite of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description. 

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.74002000000007&lon=-122.43327#.YJZJv7UzZPY')
page
```

```python
soup= bs(page.content)
print(soup.prettify())
```

```python
# Extracting day
day = soup.find_all('div', class_ = 'col-sm-2 forecast-label')
day
```

```python
Days = []
for i in day:
    Days.append(i.text)
Days
```

```python
Day = Days[0::2]
Day
```

```python
# Extracting all other details

weather= soup.find_all('div', class_ = 'col-sm-10 forecast-text')
weather
```

```python
weather_details = []
split = '.'
for i in weather: 
    weather_details.append(i.text)
weather_details
```

```python
desc=[]
split=','
for i in weather_details:
    desc.append(i.partition(split)[0])
desc
```

```python
description=desc[0::2]
description
```

```python
#temperature
split= '.'
temp = []
for i in weather_details:
    temp.append(i.partition(split)[0][-2:])
temp
```

```python
temperature=temp[0::2]
temperature
```

```python
Details=weather_details[0::2]
Details
```

### importing as a DataFrame

```python
import pandas as pd
San_fransisco_weather= pd.DataFrame({})
San_fransisco_weather['Day'] = Day
San_fransisco_weather['Short Description'] = description
San_fransisco_weather['Temperature'] = temperature
San_fransisco_weather['Description'] = Details
```
