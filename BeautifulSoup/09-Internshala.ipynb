### Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,company name, CTC, and apply date

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
page=requests.get('https://internshala.com/fresher-jobs')
page
```

```python
soup=bs(page.content)
print(soup.prettify())
```

```python
name_list=soup.find_all('div',class_='heading_4_5 profile')
name_list
```

```python
#extracting the Names
Names=[]
for name in name_list:
    Names.append(name.text.replace('\n',''))
Names
```

```python
company_list=soup.find_all('div',class_='heading_6 company_name')
company_list
```

```python
#extracting the company names
Company=[]
for company in company_list:
    Company.append(company.text.replace('\n','').strip())
Company
```

```python
Ctc=soup.find_all('div',class_='item_body')
Ctc
```

```python
salary=[]
for sal in Ctc:
    salary.append(sal.text.replace('\n','').strip())
salary
    
```

```python
CTC=salary[1::3]
CTC
```

```python
apply_date=salary[2::3]
apply_date
```

### importing as DataFrame

```python
Fresher_jobs=pd.DataFrame({})
Fresher_jobs['Job_title']=Names
Fresher_jobs['Company']=Company
Fresher_jobs['CTC']=CTC
Fresher_jobs['Date']=apply_date

```
