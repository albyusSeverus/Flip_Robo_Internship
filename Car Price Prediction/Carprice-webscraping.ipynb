```python
#importing libraries
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
from bs4 import BeautifulSoup as bs
```

```python
#importing driver
driver=webdriver.Chrome(r'C:\Users\Alby\Desktop\chromedriver\chromedriver.exe')
```

```python
#list of citys where the data is scraped
city=['https://www.cars24.com/buy-used-cars-kolkata/?itm_source=Cars24Website&itm_medium=secondary_cta',
     'https://www.cars24.com/buy-used-cars-gurgaon/?itm_source=Cars24Website&itm_medium=secondary_cta'
     'https://www.cars24.com/buy-used-cars-bengaluru/?itm_source=Cars24Website&itm_medium=secondary_cta',
     'https://www.cars24.com/buy-used-cars-chennai/?itm_source=Cars24Website&itm_medium=sticky_header',
     'https://www.cars24.com/buy-used-cars-ahmedabad/?itm_source=Cars24Website&itm_medium=secondary_cta',
     'https://www.cars24.com/buy-used-cars-noida/?itm_source=Cars24Website&itm_medium=secondary_cta'
     'https://www.cars24.com/buy-used-cars-pune/?itm_source=Cars24Website&itm_medium=sticky_header']
```

```python
#accesiing the urls of each car

car_url=[]

for lin in city:
    
    driver.get(lin)
    time.sleep(3)
    #scolling till end of the page
    for i in range(1,50):
        time.sleep(8)
        driver.execute_script("window.scrollBy(0, 1000)") 

        time.sleep(5)

   

    #accessing urls
    car=driver.find_elements_by_xpath("//a[@class='_9Ue0B']")
    for k in car:
        car_url.append(k.get_attribute('href'))
```

```python
#instances for the list
name=[]
history=[]
owner=[]
km=[]
engine=[]
transmission=[]
insurance=[]
purchase=[]
price=[]
lservice=[]
i=1


#accessing each url of the car

for link in car_url:
    print(f"url no {i}")
    try:
        time.sleep(3)
        driver.get(link)
    except:
        time.sleep(30)
        driver.get(link)
        
        
        #name of the car
        
    try:
        nam=driver.find_element_by_xpath("//h2[@class='_2geSF']")
        name.append(nam.text)
    except:
         name.append('-')
            
            
            
            #price of the car
            
            
    try:        
        pri=driver.find_element_by_xpath("//strong[@class='_2yYvS']")
        price.append(pri.text)
    except:
        price.append('-')
        
        
        #accessing entire table and sorting them based on their label
        
    tabs=[]
    
    try:
        
        tab=driver.find_elements_by_xpath("//li[@class='tHlIu']")
        for el in tab:
            tabs.append(el.text)

        for j in tabs:

            if 'Transmission' in j:
                transmission.append(j)

            if 'Owner' in j:
                owner.append(j)

            if 'History' in j:
                history.append(j)

            if 'Year of Purchase' in j:
                purchase.append(j)

            if 'Kilometers Driven' in j:
                km.append(j)

            if 'Last Service' in j:
                lservice.append(j)

            if 'Fuel Type' in j:
                engine.append(j)
                
            if 'Insurance' in j:
                insurance.append(j)
    
    except:
        
        transmission.append('-')
        owner.append('-')
        history.append('-')
        purchase.append('-')
        km.append('-')
        lservice.append('-')
        engine.append('-')
        insurance.append('-')
          
    i=i+1
    
    
   



            
        
        
```

```python
#assigning the list to a new list for data cleaning
historyt=history
ownert=owner
kmt=km
enginet=engine
transmissiont=transmission
insurancet=insurance
purchaset=purchase
lservicet=lservice
```

```python
# marking the indices where the data was not scraped

for j in indices:
    historyt.insert(j,'-')
    transmissiont.insert(j,'-')
    ownert.insert(j,'-')
    purchaset.insert(j,'-')
    kmt.insert(j,'-')
    lservicet.insert(j,'-')
    enginet.insert(j,'-')
    insurancet.insert(j,'-')
```

```python
#Extracting useful information from the collected data, cleaning the data


transmissionf=[]
transmissionf1=[]
insurancetype=[]
ownerf=[]
purchasef=[]
enginef=[]
insurancef=[]
insurancefyear=[]
kmf=[]
makeyear=[]
brand=[]
model=[]
pricef=[]

for i in transmissiont:
    if 'Transmission' in i:
        transmissionf.append(i.replace('Transmission',''))
    else:
        transmissionf.append(i)
for j in transmissionf:
    if '\n' in j:
        transmissionf1.append(j.replace('\n',''))
    else:
        transmissionf1.append(j)
for k in ownert:
    if 'Owner\n' in k:
        ownerf.append(k.replace('Owner\n',''))
    else:
        ownerf.append(k)
        
for l in purchaset:
    if 'Year of Purchase\n' in l:
        purchasef.append(l.replace('Year of Purchase\n',''))
    else:
        purchasef.append(l)
        
for m in enginet:
    if 'Fuel Type\n' in m:
        enginef.append(m.replace('Fuel Type\n',''))
    else:
        enginef.append(m)

for n in insurancet:
    if 'Insurance\nValid upto' in n:
        insurancef.append(n.replace('Insurance\nValid upto',''))
    else:
        insurancef.append(n)
for o in insurancef:
    if '\n' in o:
        insurancefyear.append(o.split('\n')[0])
        insurancetype.append(o.split('\n')[1])
    else:
        insurancefyear.append(o)
        insurancetype.append(o)
        
for p in kmt:
    if 'Kilometers Driven\n' in p:
        kmf.append(p.replace('Kilometers Driven\n',''))
    else:
        kmf.append(p)
        
for q in name:
    if q!='-':
        makeyear.append(q.split(' ')[0])
        brand.append(q.split(' ')[1])
        model.append(q.split(' ')[2])
    else:
        makeyear.append(q)
        brand.append(q)
        model.append(q)
for r in price:
    if r!='-':
        k=r.split('')
        if len(k)==1:
            pricef.append(r.split('')[1])
        else:
            pricef.append(r.split('')[-1])
    else:
        pricef.append(r)
        
    
        
        
    
```

```python
#removing all the empty values
newname=[]
newmodel=[]
newbrand=[]
newhis1=[]
newtrans=[]
neweng=[]
newkm=[]
newmakeyear=[]
newinsurancetype=[]
newpurchase=[]
newinsyear=[]
newprice=[]

for i in name:
    if i!='-':
        newname.append(i)
for j in model:
    if j!='-':
        newmodel.append(j)
for k in brand:
    if k!='-':
        newbrand.append(k)
for l in newhis:
    if l!='none':
        newhis1.append(l)
for m in transmissionf1:
    if m!='-':
        newtrans.append(m)
for n in enginef:
    if n!='-':
        neweng.append(n)
for o in kmf:
    if o!='-':
        newkm.append(o)
        
for p in makeyear:
    if p!='-':
        newmakeyear.append(p)
        
for q in insurancetype:
    if q!='-':
        newinsurancetype.append(q)
for r in purchasef:
    if r!='-':
        newpurchase.append(r)
for s in insurancefyear:
    if s!='-':
        newinsyear.append(s)
for t in pricef:
    if t!='-':
        newprice.append(t)
```

```python
#cleaning additional info from lists

newhis2=[]
newkm1=[]
instype3=[]
newpurchase2=[]
newinsyear2=[]
owner3=[]
owner4=[]

for his in newhis1:
    if 'History\n' in his:
        newhis2.append(his.replace('History\n',''))
    else:
        
        newhis2.append(his)
        
for kil in newkm:
    if 'km' in kil:
        newkm1.append(kil.replace('km',''))
    else:
        newkm1.append(kil)
for ins in newinsurancetype:
    if ins=='Third_party':
        instype3.append(ins.replace('Third_party','3rd Party'))
    elif ins=='Third Party':
        instype3.append(ins.replace('Third Party','3rd Party'))
    elif ins=='Comprehensive':
        instype3.append(ins.replace('Comprehensive','Comp'))
    elif ins=='Zero Depreciation':
        instype3.append(ins.replace('Zero Depreciation','Zero_Dep'))
    else:
        instype3.append(ins)
        


for year in newpurchase:
    newpurchase2.append(year.split(' ')[1])
    
for kollam in newinsyear:
    newinsyear2.append(kollam.split(' ')[2])

for ow in ownerf:
    if ow!='-':
        owner3.append(ow)
for m in owner3:
    if m=='Owner1st Owner':
        owner4.append(m.replace('Owner1st Owner','1st Owner'))
    elif m=='Owner2nd Owner':
        owner4.append(m.replace('Owner2nd Owner','2nd Owner'))
        
    elif m=='Owner3rd Owner':
        owner4.append(m.replace('Owner3rd Owner','3rd Owner'))
    else:
        owner4.append(m)
        
```

```python
#creating dataframe

df=pd.DataFrame({})
df['Name']=newname
df['Model']=newmodel
df['Brand']=newbrand
df['No_pre_owners']=owner4
df['accident_history']=newhis2
df['Transmission']=newtrans
df['Engine']=neweng
df['KM_used']=newkm1
df['Manufacture year']=newmakeyear
df['Insurance_type']=instype3
df['Purchase_year']=newpurchase2
df['Insurance_valid_till']=newinsyear2
df['price']=newprice
df
```

```python
#saving the csv file

df.to_csv('used_cars.csv')
```
