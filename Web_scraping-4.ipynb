### 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

```python
#getting url
driver.get('https://github.com/')
time.sleep(4)

#getting trending button

dr = driver.find_elements_by_xpath("//li[@class='edge-item-fix']/a")
dra = []
for i in dr:
    dra.append(i.get_attribute('href'))

for i in dra:
    if ('trending' in i):
        driver.get(i)
```

```python
#respository name

repo = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
repos = []
for i in repo:
    repos.append(i.text)
    
#description

des = driver.find_elements_by_xpath("//p[@class = 'col-9 color-text-secondary my-1 pr-4']")
desx = []
for i in des:
    desx.append(i.text)

#count

cont = driver.find_elements_by_xpath("//a[@class = 'Link--muted d-inline-block mr-3']")
contx = []
for i in cont:
    contx.append(i.text)
contributor_count = contx[1::2]

#language

language=[]
lang=driver.find_elements_by_xpath("//span[@class='d-inline-block ml-0 mr-3']")
for i in lang:
    if i.text is None :
        language.append("--") 
    else:
        language.append(i.text)
```

```python

Git = pd.DataFrame({})
Git['Repos'] = repos[:21]
Git['Description'] = desx[:21]
Git['Count'] = contributor_count[:21]
Git['language'] = language[:21]
Git
```

### 9. Scrape the details most watched tv series of all time from imdb.com.
 Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes


```python
#getting url

driver.get('https://www.imdb.com/list/ls095964455/')
time.sleep(5)
```

```python
#getting details


#name

name = []
nm = driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in nm:
    name.append(i.text)

    #year
    
year = []
yr = driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in yr:
    year.append(i.text)
  

    #genre 
    
    
genre = []
gen = driver.find_elements_by_xpath("//span[@class='genre']")
for i in gen:
    genre.append(i.text)
    
run = []
rn = driver.find_elements_by_xpath("//span[@class='runtime']")
for i in rn:
    run.append(i.text)

    #rationg


rating = []
rat = driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']")
for i in rat:
    rating.append(i.text.replace('\nRate', ''))

    
    #votes


votes = []
vt = driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]")
for i in vt:
    votes.append(i.text.replace('Votes: ',''))
```

```python
#creating dataframe
IMDB = pd.DataFrame({})
IMDB['Name'] = name
IMDB['Year'] = year
IMDB['Ratings'] = rating
IMDB['Genre'] = genre
IMDB['Runtime'] = run
IMDB['Vote'] = votes
IMDB
```
