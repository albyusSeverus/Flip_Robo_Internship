## 07- Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
page=requests.get('https://www.amazon.in/s?k=mobile+phones+under+20000&i=electronics&ref=nb_sb_noss_1')
page
```

```python
soup=bs(page.content)
print(soup.prettify())
```

```python
#finding names
name_list=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal',limit=22)
name_list
```

```python
Model=[]
split='('
for name in name_list:
    Model.append(name.text.partition(split)[0].strip())
Model
```

```python
#taking the prices
prices = soup.find_all('span', class_ = 'a-price-whole',limit=22)
prices
```

```python
Price = []
for i in prices: 
    Price.append(i.text.lstrip())
Price
```

```python
#taking out the urls
img_urls = soup.find_all('img', class_ = 's-image',limit=22)
img_urls
```

```python
URL = []
for i in img_urls:
    URL.append(i.get('src'))
URL
```

```python
# rating
rating = soup.find_all('span', class_ = 'a-icon-alt',limit=22)
rating
```

```python
Rating=[]
for i in rating:
    Rating.append(i.text.split()[0])
Rating
```

### Importing as DataFrame

```python
Phones_under_20k = pd.DataFrame({})
Phones_under_20k['Phone Name'] = Model
Phones_under_20k['Price'] = Price
Phones_under_20k['Rating'] = Rating
Phones_under_20k['Image URL'] = URL
```
