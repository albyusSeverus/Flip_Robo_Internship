```python
from bs4 import BeautifulSoup
import requests
```

### All the header tags from ‘en.wikipedia.org/wiki/Main_Page

```python
def wiki_header(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    all_h=soup.find_all('h2',class_='mp-h2')
    headers=[]
    for i in all_h:
        headers.append(i.text.replace('\xa0',''))
    print(headers)
    
    
```

```python
wiki_header('https://en.wikipedia.org/wiki/Main_Page')
```

### IMDB’s Top rated 100 movies

```python
import pandas as pd
Imdb_top100=pd.DataFrame({})
def imdb(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    all_movies=soup.find_all('td',class_='titleColumn',limit=100)
    all_rating=soup.find_all('td',class_='ratingColumn imdbRating',limit=100)
    movie_name=[]
    movie_year=[]
    movie_rating=[]
    for i in all_movies:
        for j in i.find_all('a'):
            movie_name.append(j.text.replace('\n',''))
        for year in i.find_all('span'):
            movie_year.append(year.text.strip('()'))
    for rating in all_rating:
        movie_rating.append(rating.text.replace('\n',''))
    Imdb_top100['Movie_Name']=movie_name
    Imdb_top100['Year']=movie_year
    Imdb_top100['Rating']=movie_rating
```

```python
imdb('https://www.imdb.com/chart/top/')
```

### 5 books from ‘www.bookpage.com’


```python
Books=pd.DataFrame({})
def bookreview(url):
    book_name=[]
    Genre=[]
    Writer=[]
    Review=[]
    page=requests.get(url)
    books=BeautifulSoup(page.content)
    title=books.find_all('div',class_='flex-article-content',limit=5)
    for i in title:
        for name in i.find_all('h4',class_='italic'):
            book_name.append(name.text.replace('\n','').lstrip().strip('★ '))
        for writr in i.find_all('p',class_='sans bold'):
            Writer.append(writr.text.replace('\n','').lstrip().strip('★ '))
        for genre in i.find_all('p',class_='genre-links hidden-phone'):
            Genre.append(genre.text.replace('\n',''))
        for review in i.find_all('p',class_='excerpt'):
            Review.append(review.text.replace('\n',''))
            
    Books['Name']=book_name
    Books['Authors']=Writer
    Books['Genre']=Genre
    Books['Review']=Review
            
    
```

```python
bookreview('https://bookpage.com/reviews')
Books
```
